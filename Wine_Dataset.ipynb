```python
import pandas as pd
from sklearn.datasets import load_wine

# Load the Wine dataset
data = load_wine()

# Create a pandas DataFrame from the dataset
df = pd.DataFrame(data.data, columns=data.feature_names)

# Add the target variable to the DataFrame
df['target'] = data.target

# View the resulting DataFrame
df.head()
```

```python
new_data = df.sample(frac=1, random_state=42)
new_data.head(15)
```

```python
features = new_data.drop(['target'], axis=1)
```

```python
label = new_data['target']
```

```python
features.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features_new = scaler.fit_transform(features)
```

```python
features_new
```

```python
features_new1= np.insert(features_new, 0, 1, axis=1)
features_new1.shape
```

```python
val = list(label.unique())
val
```

```python
one_hot_encoded_data = np.asarray(pd.get_dummies(label, columns = ['target']))
label = one_hot_encoded_data
```

```python
label
```

```python
features_new1.shape
```

```python
label.shape
```

```python
train_x = features_new1[:100, :]
train_y = label[:100, :]
val_x = features_new1[100:, :]
val_y = label[100:, :]
```

```python
alpha = 0.01
np.random.seed(42)
weights = np.random.rand(15,3)
loss_list = []
val_loss = []
for i in range(1000):
  z= np.dot(train_x,weights)
  num = np.exp(z)
  den = np.sum(num, axis=1)
  print(num.shape)
  print(den.shape)
  den = den.reshape(den.shape[0],1)
  print(den.shape)
  y_hat = num/den
  loss = -np.mean(np.sum(train_y*np.log(y_hat), axis=1))
  print('Epoch: '+str(i+1)+" Loss: "+str(loss))
  loss_list.append(loss)
  dw = 1/train_x.shape[0] * (np.dot(train_x.T, (y_hat - train_y) ))
  weights-=alpha * dw

  # Validation
  z = np.dot(val_x,weights)
  num = np.exp(z)
  den = np.sum(num, axis=1)
  den = den.reshape(den.shape[0],1)
  y_hat = num/den
  loss = -np.mean(np.sum(val_y*np.log(y_hat), axis=1))
  print("Val Loss: "+str(loss))
  val_loss.append(loss)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(len(loss_list)), loss_list)
plt.plot(range(len(val_loss)), val_loss)
plt.show()
```
